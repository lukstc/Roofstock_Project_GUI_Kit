```python
import pandas as pd
```

```python
dt = pd.read_csv('list_offer_full_Dec05.csv')
```

```python
dt_1 = dt.groupby('LISTING_ID').ffill()
dt_1 = dt_1.groupby('LISTING_ID').bfill()
```

```python
# set working bar
#GOOD_OFFER_BAR = 0.95
#GOOD_OFFER_BAR = 0.90
#GOOD_OFFER_BAR = 0.85
#GOOD_OFFER_BAR = 0.80
#GOOD_OFFER_BAR = 0.75
GOOD_OFFER_BAR = 0.70
```

```python
dt_1['GOOD_OFFER'] = (dt_1['OFFER_PRICE'] >= dt_1['LIST_PRICE'] * GOOD_OFFER_BAR)
```

```python
to_drop_index = dt_1[(dt_1['RECORD_TYPE']=="Offer") & (dt_1['GOOD_OFFER']==False)].index
```

```python
# dfObj.drop(indexNames , inplace=True)
dt_1.drop(to_drop_index,inplace=True)
```

```python
# dt_1.head(50)
# dt_1.isna().sum()
```

```python
dt_1.groupby('RECORD_STATUS').count()['LISTING_ID']
```

```python
def new_value(est,v_low,v_high):
    if pd.isnull(est):
        return (v_high+v_low)/2
    else:
        return est
```

```python
dt_1['EST_VALUATION'] = dt_1.apply(lambda x: new_value(x.EST_VALUATION,x.EST_VALUATION_LOW,x.EST_VALUATION_HIGH),axis=1)
```

```python
dt_1[dt_1.LISTING_ID==1631771].groupby(['LISTING_ID','RECORD_TYPE','RECORD_STATUS','LIST_PRICE'])['RECORD_TS'].min()
```

```python
grouped = dt_1.groupby(['LISTING_ID','RECORD_TYPE','RECORD_STATUS','LIST_PRICE'])['RECORD_TS'].min()
```

```python
dict(grouped)
```

```python
dt_2 = [] # dt_2 save the running flag
for k,v in dict(grouped).items():
    #if 'For Sale' in  k[2] or 'Off' in  k[2] or 'New' in k[2] or 'Counter' in k[2]:
    dt_2.append((v,k[0],k[1],k[2],k[3]))
```

```python
def get_key(tuple_item):
    return tuple_item[0]
dt_2 = sorted(dt_2,key=get_key)
```

```python
test = {}
for item in dt_2:
    if item[3] in list(test.keys()):
        test[item[3]]+=1
    else:
        test[item[3]]=1
```

```python
result_list = []
len_dt_2 = len(dt_2)

for i in range(len_dt_2):
    
    current = dt_2[i]
    print(i/len_dt_2*100)
    # current[0] time-stamp
    # current[1] list-id
    # current[2] record-type
    # current[3] record-status
    # current[4] list-price
    
    new_rec = {}
    if 'For Sale' in current[3]:
        new_rec['LIST_ID'] = current[1]
        new_rec['START_TS'] =current[0]
        new_rec['END_TS'] = None
        new_rec['LIST_PRICE'] = current[4]
        new_rec['EST_VALUE'] = dt_1.loc[(dt_1['LISTING_ID']==current[1]) & (dt_1['RECORD_TS']==current[0]),'EST_VALUATION'].values[0]
        new_rec['OFFER_PRICE'] = None
        new_rec['EVENT'] = False
        new_rec['ISFEATURED']=dt_1.loc[(dt_1['LISTING_ID']==current[1]) & (dt_1['RECORD_TS']==current[0]),'ISFEATURED'].values[0]
        new_rec['OCCUPANCY']=dt_1.loc[(dt_1['LISTING_ID']==current[1]) & (dt_1['RECORD_TS']==current[0]),'OCCUPANCY'].values[0]
        new_rec['ISRENTGUARANTEED']=dt_1.loc[(dt_1['LISTING_ID']==current[1]) & (dt_1['RECORD_TS']==current[0]),'ISRENTGUARANTEED'].values[0]
        new_rec['BEDROOMS']= dt_1.loc[(dt_1['LISTING_ID']==current[1]) & (dt_1['RECORD_TS']==current[0]),'BEDROOMS'].values[0]
        new_rec['BATHROOMS']= dt_1.loc[(dt_1['LISTING_ID']==current[1]) & (dt_1['RECORD_TS']==current[0]),'BATHROOMS'].values[0]
        new_rec['NEIGHBORHOODSCORE']= dt_1.loc[(dt_1['LISTING_ID']==current[1]) & (dt_1['RECORD_TS']==current[0]),'NEIGHBORHOODSCORE'].values[0]
        new_rec['SCHOOLHIGHSCORE']= dt_1.loc[(dt_1['LISTING_ID']==current[1]) & (dt_1['RECORD_TS']==current[0]),'SCHOOLHIGHSCORE'].values[0]
        new_rec['SCHOOLMIDDLESCORE']= dt_1.loc[(dt_1['LISTING_ID']==current[1]) & (dt_1['RECORD_TS']==current[0]),'SCHOOLMIDDLESCORE'].values[0]
        new_rec['ISEXCLUSIVE']= dt_1.loc[(dt_1['LISTING_ID']==current[1]) & (dt_1['RECORD_TS']==current[0]),'ISEXCLUSIVE'].values[0]
        new_rec['COMPUTED_CAP_RATE']= dt_1.loc[(dt_1['LISTING_ID']==current[1]) & (dt_1['RECORD_TS']==current[0]),'COMPUTED_CAP_RATE'].values[0]
        new_rec['COMPUTED_GROSS_YIELD']= dt_1.loc[(dt_1['LISTING_ID']==current[1]) & (dt_1['RECORD_TS']==current[0]),'COMPUTED_GROSS_YIELD'].values[0]
        new_rec['COMPUTEDLEVEREDCASHONCASH']= dt_1.loc[(dt_1['LISTING_ID']==current[1]) & (dt_1['RECORD_TS']==current[0]),'COMPUTEDLEVEREDCASHONCASH'].values[0]
        new_rec['COMPUTEDTOTALRETURN']= dt_1.loc[(dt_1['LISTING_ID']==current[1]) & (dt_1['RECORD_TS']==current[0]),'COMPUTEDTOTALRETURN'].values[0]
        new_rec['EST_REPAIR_COST']= dt_1.loc[(dt_1['LISTING_ID']==current[1]) & (dt_1['RECORD_TS']==current[0]),'EST_REPAIR_COST'].values[0]
        new_rec['LEASEMONTHLYRENT']= dt_1.loc[(dt_1['LISTING_ID']==current[1]) & (dt_1['RECORD_TS']==current[0]),'LEASEMONTHLYRENT'].values[0]
        new_rec['INSPECTION_TYPE_ID']= dt_1.loc[(dt_1['LISTING_ID']==current[1]) & (dt_1['RECORD_TS']==current[0]),'INSPECTION_TYPE_ID'].values[0]
        for j in range(i+1,len_dt_2):
            candi = dt_2[j]
            if candi[2] == "List":
                if (
                    candi[1] == current[1] # same id
                    and 'For Sale' in candi[3]
                    and current[4] != candi[4]
                ):
                    new_rec['END_TS'] = candi[0] # End-date
                    break
                elif (
                    candi[1] == current[1] # same id
                    and 'Off Market' in candi[3]
                ):
                    new_rec['END_TS'] = candi[0] # End-date
                    break
                else:
                    pass
            else: # offer
                if (
                    candi[1] == current[1] # same list-id
                    and ('New' in candi[3] or 'Countered' in candi[3] or 'Accepted' in candi[3]) # valid offer
                    and candi[4] == current[4] # same list Price
                ):
                    new_rec['END_TS'] = candi[0] 
                    new_rec['OFFER_PRICE'] = dt_1.loc[(dt_1['LISTING_ID']==candi[1]) & (dt_1['RECORD_TS']==candi[0]),'OFFER_PRICE'].values[0]
                    new_rec['EVENT']= True
                    break
                    
        result_list.append(new_rec)
    else:
        pass
```
